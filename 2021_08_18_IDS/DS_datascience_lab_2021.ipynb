```python
df = pd.read_csv('olympics.csv', index_col = 0 , skiprows = 1)

for col in df.columns:
    if col[:2]=='01':
        df.rename(columns={col:'Gold'+col[4:]}, inplace=True)
    if col[:2]=='02':
        df.rename(columns={col:'Silver'+col[4:]}, inplace=True)
    if col[:2]=='03':
        df.rename(columns={col:'Bronze'+col[4:]}, inplace=True)
    if col[:1]=='№':
        df.rename(columns={col:'#'+col[4:]}, inplace=True)
        
names_ids = df.index.str.split('\s\(')
        
df.index = names_ids.str[0]
df['ID'] = names_ids.str[1].str[:3]

df = df.drop('Totals')
```

```python

#Question # 6

def answer_seven():
    stat = df1[df1['SUMLEV']==50]
    pop_stat = ["POPESTIMATE2010","POPESTIMATE2011","POPESTIMATE2012","POPESTIMATE2013","POPESTIMATE2014","POPESTIMATE2015"]
    return stat.iloc[(stat[pop_stat].max(axis=1)-stat[pop_stat].min(axis=1)).argmax()]['CTYNAME']
```

```python
answer_seven()
```

```python
def answer_eight():
    pop_stat = df1[df1['SUMLEV']==50]
    region = pop_stat[(pop_stat['REGION']==1) | (pop_stat['REGION']==2)]
    washington = region[region['CTYNAME'].str.startswith("Washington")]
    popestimate = washington[washington['POPESTIMATE2015'] > washington['POPESTIMATE2014']]
    return popestimate[['STNAME','CTYNAME']]
```

```python
answer_eight()
```
