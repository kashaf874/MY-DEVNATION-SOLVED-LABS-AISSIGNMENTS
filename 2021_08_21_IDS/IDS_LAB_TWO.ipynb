```python
def answer_one():
    df = pd.ExcelFile('Energy_Indicators.xls')
    my_doc = df.parse(skiprows=17,skip_footer=(38))
    
    #replace values with new val_ acc to question requirement
    
    my_doc = my_doc[['Unnamed: 1','Petajoules','Gigajoules','%']]
    my_doc.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
    my_doc[['Energy Supply', 'Energy Supply per Capita', '% Renewable']] =  my_doc[['Energy Supply', 'Energy Supply per Capita', '% Renewable']].replace('...',np.NaN).apply(pd.to_numeric)
    
    #convertung E supply to gigajoules
    
    my_doc['Energy Supply'] = my_doc['Energy Supply']*1000000
    my_doc['Country'] = my_doc['Country'].replace({'China, Hong Kong Special Administrative Region':'Hong Kong','United Kingdom of Great Britain and Northern Ireland':'United Kingdom','Republic of Korea':'South Korea','United States of America':'United States','Iran (Islamic Republic of)':'Iran'})

    
    #renaming of the respacted countries above and
    # skiping header here
    
    GDP = pd.read_csv('world_bank.csv',skiprows=4)
    GDP['Country Name'] = GDP['Country Name'].replace('Korea, Rep.','South Korea')
    GDP['Country Name'] = GDP['Country Name'].replace('Iran, Islamic Rep.','Iran')
    GDP['Country Name'] = GDP['Country Name'].replace('Hong Kong SAR, China','Hong Kong')
    
    
    GDP = GDP[['Country Name','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015']]
    GDP.columns = ['Country','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015']

    ScimEn = pd.read_excel('scimagojr-3.xlsx')
    ScimEn_one = ScimEn[:15]
    
    df = pd.merge(ScimEn_m,my_doc,how='inner',left_on='Country',right_on='Country')
    result_df = pd.merge(df,GDP,how='inner',left_on='Country',right_on='Country')
    result_df = result_df.set_index('Country')
    
    return result_df
```

```python
def answer_two():
    Energy = pd.read_excel('Energy_Indicators.xls',na_values=["..."],header = None,skiprows=18,skipfooter= 38,names=['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable'])
    Energy['Energy Supply'] = Energy['Energy Supply']*1000000


    Energy['Country'] = Energy['Country'].replace({'Republic of Korea' : 'South Korea','United States of America' : 'United States','United Kingdom of Great Britain and Northern Ireland':'United Kingdom','China, Hong Kong Special Administrative Region':'Hong Kong'})

    GDP = pd.read_csv('world_bank.csv', skiprows = 4)
    GDP['Country Name'] = GDP['Country Name'].replace({'Korea, Rep.': 'South Korea','Iran, Islamic Rep.': 'Iran','Hong Kong SAR, China' : 'Hong Kong'})

    ScimEn = pd.read_excel('scimagojr-3.xlsx')

    inner_one = pd.merge(ScimEn,Energy,how="inner",left_on="Country",right_on="Country")

    GDP.rename(columns = {"Country Name":"Country"},inplace=True)
    GDP = GDP.loc[:,['2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015',"Country"]]
    inner_two = pd.merge(inner_one,GDP,how="inner",left_on="Country",right_on="Country").set_index("Country")

    outer_one = pd.merge(ScimEn,Energy,how="outer",left_on="Country",right_on="Country")
    outer_two = pd.merge(outer_one,GDP,how="outer",left_on="Country",right_on="Country").set_index("Country")

    return len(outer_one)-len(inner_two)

```

```python
#Question three

def answer_three():
    Top_countries = answer_one()
    avg_GDP = Top_countries[['2006','2007','2008','2009','2010','2011','2012','2013','2014','2015']].mean(axis=1).rename('avgGDP').sort_values(ascending=False)
    return avg_GDP

answer_three()
```

```python
#Q__NINE

def answer_nine():
    info = answer_one()
    info['PopEst'] = info['Energy Supply'] / info['Energy Supply per Capita']
    info['Citable docs per Capita'] = info['Citable documents'] / info['PopEst']
    ans = info['Citable docs per Capita'].corr(info['Energy Supply per Capita'])
    return ans

answer_nine()
```

```python
#Q__TEN

def answer_ten():
    info = answer_one()
    Rmedian=info["% Renewable"].median()
    info["HighRenew"]= info["% Renewable"].apply(lambda x:0 if x<Rmedian else 1 )
    return info["HighRenew"]
```

```python
#Q___11

def answer_eleven():
    Top15 = answer_one()
    ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
    Top15["popu"]=Top15['Energy Supply']/Top15['Energy Supply per Capita']
    Top15["popu"]=Top15["popu"]
    Top15["Continent"] = pd.Series(ContinentDict)
    return Top15.groupby("Continent")["popu"].agg([np.size,np.sum, np.mean, np.std])

```
