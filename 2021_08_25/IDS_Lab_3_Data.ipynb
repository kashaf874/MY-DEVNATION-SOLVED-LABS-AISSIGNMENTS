```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('nobel.csv')
df.head(50)
```

```python
#Question # 2
def answer_first():
     return df['birth_date'].value_counts().idxmax(), df['sex'].value_counts().idxmax()


    
```

```python
answer_first()
```

```python
#Question # 3

def answer_third():
    df['USA_born_winner'] = df['birth_country']=='United States of America'
    df['decade'] = (np.floor(df['year']/10)*10).astype(int)
    prop_usa_winners = df.groupby('decade',as_index=False)['USA_born_winner'].mean()
    return prop_usa_winners
```

```python
answer_third()
```

```python
#Question 4
plt.rcParams['figure.figsize'] = [9, 6]
def answer_four():
    return sns.lineplot(x='decade', y='USA_born_winner', data=df)

```

```python
answer_four()
```

```python
#Question 5

def answer_five():
    df['female_winner'] = df["sex"] == "Female"
    female_data = df.groupby(["decade", "category"], as_index=False)["female_winner"].mean()
    return sns.lineplot(x="decade", y="female_winner", hue="category",data= female_data)
```

```python
answer_five()
```

```python

df['birth_date'] = pd.to_datetime(df["birth_date"])
def answer_eight():
    # Calculating the age of Nobel Prize winners
    df['age'] = df["year"] - df["birth_date"].dt.year
    return sns.lmplot(x="year", y="age", data=df, lowess=True, aspect=2, line_kws={"color" : "black"})
```

```python
answer_eight()
```
